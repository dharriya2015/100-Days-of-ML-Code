```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
# initialize setup
from pycaret.regression import *
# s = setup(df, target = 'Survived', transform_target = True, log_experiment = True, experiment_name = 'Titanic')
s = setup(df, target = 'Survived')
```

```python
best = compare_models()
```

```python
# check the final params of best model
best.get_params()
```

```python
# check the residuals of trained model
plot_model(best, plot = 'residuals_interactive')
```

```python
evaluate_model(best)
```

```python
df_test=pd.read_csv('test.csv')
df_test.head()
```

```python
best
```

```python
pred = predict_model(best, data = df_test)
```

```python
pred
```

```python
results = pred.copy()
results[['prediction_label']].median()
```

```python
results
```

```python
med = results['prediction_label'].median()
```

```python
results['Survived'] = (results['prediction_label'] >= med).astype(int)
results = results[['PassengerId', 'Survived']]
results.to_csv("submission_pyc_med.csv",index = False)
```
